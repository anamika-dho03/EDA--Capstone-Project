## 3. ***Data Wrangling***

````python
print(df.columns)
````

````python
df.columns = df.columns.str.strip()
````

````python
# Data Wrangling Code

# 1. Select only important columns
df = df[['iyear','country_txt','region_txt','city',
         'attacktype1_txt','weaptype1_txt','targtype1_txt',
         'gname','nkill','nwound',]]

# 2. Handle missing values
df.loc[:, 'nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)

df['country_txt'] = df['country_txt'].fillna("Unknown")
df['city'] = df['city'].fillna("Unknown")
df['attacktype1_txt'] = df['attacktype1_txt'].fillna("Unknown")
df['weaptype1_txt'] = df['weaptype1_txt'].fillna("Unknown")
df['targtype1_txt'] = df['targtype1_txt'].fillna("Unknown")
df['gname'] = df['gname'].fillna("Unknown")

# 3. Remove duplicate rows
df = df.drop_duplicates()

# 4. Create a new column for total casualties
df['casualties'] = df['nkill'] + df['nwound']

# 5. Final dataset preview
df.head()
````

### What all manipulations have you done and insights you found?



## Data Wrangling Summary & Insights

- Selected only relevant columns to simplify analysis
- Handled missing values by replacing them with appropriate values (0 or "Unknown")
- Removed duplicate rows to improve data quality
- Created a new column 'casualties' by combining killed and wounded counts

### Insights:
- Dataset contains many missing values which were handled
- Data cleaning helps in making analysis more accurate
- New column 'casualties' helps measure total impact of attacks

#### Chart - 14 - Correlation Heatmap

````python
# Chart - 14: Correlation Heatmap

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

# Select only numeric columns
corr = df[['nkill','nwound','casualties']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap of Casualties")

plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used to understand the relationship between numerical variables. It helps in identifying how strongly variables like number of kills, wounded, and total casualties are related to each other.

##### 2. What is/are the insight(s) found from the chart?

 The heatmap shows a strong positive correlation between number of kills, number of wounded, and total casualties. This indicates that as the number of people killed increases, the number of people wounded also tends to increase.

#### Chart - 15 - Pair Plot

````python
# Chart - 15: Pair Plot

import seaborn as sns
import matplotlib.pyplot as plt

sample_df = df[['nkill','nwound','casualties']].sample(1000)

sns.pairplot(sample_df)

plt.suptitle("Pair Plot of Casualties Variables", y=1.02)

plt.show()
````

A pair plot is used to visualize relationships between multiple numerical variables at once. It helps in understanding distributions as well as pairwise relationships between variables.

##### 2. What is/are the insight(s) found from the chart?

The pair plot shows positive relationships between number of kills, wounded, and total casualties. It also shows that most data points are concentrated at lower values, indicating that most attacks result in fewer casualties.
